```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
url_1 = "https://profiles.ucr.edu/api/profile?acctStruct=&groupId=1831&researchArea=undefined&affiliationFilter=All&excludeStudentEmployees=false&excludeSecondaryDepartmentEmployees=false&excludeAffiliates=false"
url_2 = "https://profiles.ucr.edu/api/profile?acctStruct=&groupId=1787&researchArea=undefined&affiliationFilter=All&excludeStudentEmployees=false&excludeSecondaryDepartmentEmployees=false&excludeAffiliates=false"
url_3 = "https://profiles.ucr.edu/api/profile?acctStruct=&groupId=1809&researchArea=undefined&affiliationFilter=All&excludeStudentEmployees=false&excludeSecondaryDepartmentEmployees=false&excludeAffiliates=false"
url_4 = "https://profiles.ucr.edu/api/profile?acctStruct=&groupId=1820&researchArea=undefined&affiliationFilter=All&excludeStudentEmployees=false&excludeSecondaryDepartmentEmployees=false&excludeAffiliates=false"
```

```python
urls = [url_1, url_2, url_3, url_4]
```

```python
def faculty_ucr_cs(url):
    
    response = requests.get(url)
    result = response.json()
    return result
    

```

```python
all_results = []

for url in urls:
    result = faculty_ucr_cs(url)
    all_results.extend(result)
    time.sleep(5)
```

```python
df = pd.DataFrame(all_results) 
```
